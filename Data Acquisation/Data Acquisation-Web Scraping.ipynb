```python
from urllib.request import urlopen
```

```python
wiki_url="https://en.wikipedia.org/wiki/Indian_rupee"
```

```python
wiki_data=urlopen(wiki_url)
print(type(wiki_data))
```

```python
wiki_html=wiki_data.read()
print(wiki_html)
wiki_data.close()
```

# Parsing data

```python
from bs4 import BeautifulSoup as soup
```

```python
wiki_soup=soup(wiki_html,'html.parser')
print(wiki_soup)
```

```python
wiki_soup.h1
```

```python
wiki_soup.findAll('h1',{})
```

```python
tables=wiki_soup.findAll('table',{'class':'wikitable'})
print(len(tables))
```

```python
ruppee_table=tables[1]
# print(ruppee_table)
```

# Parsing HTML Tables

```python
headers=ruppee_table.findAll('th')
print(headers)
```

```python
# headers[0].text
Column_titles=[ct.text[:-1] for ct in headers]
Column_titles
```

```python
rows_data=ruppee_table.findAll('tr')[1:]
print(len(rows_data))
```

```python
first_row=rows_data[0].findAll('td',{})
# print(first_row)
for d in first_row:
    print(d.text[:-1])
```

```python
# We need to do this for all the rows
table_rows=[]
rows_data=rows_data[:-1]
for row in rows_data:
    current_row=[]
    row_data=row.findAll('td',{})
    for idx,data in enumerate(row_data):
        current_row.append(data.text[:-1])
        
    table_rows.append(current_row)
print(table_rows)
```

# Writing and reading CSV

```python
filename='rupee_history.csv'
with open(filename,'w',encoding='utf-8') as f:
    header_string=','.join(Column_titles)
    header_string+='\n'
    f.write(header_string)
    
    for row in table_rows:
        row_string=""
        row_string=",".join(row)
        row_string += '\n'
        f.write(row_string)
```
